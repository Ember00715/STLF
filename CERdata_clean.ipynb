```python
import pandas as pd
```

```python
df = pd.read_csv('File1.txt', sep=' ', names=['Line', 'Date', 'Power'])
pd.DataFrame(df)
```

```python
df.isnull().sum().sort_values(ascending=False)
df['Date'] = df['Date'].astype('string')
print(df.dtypes)
```

```python
df['Days'] = [x[3:5] for x in df['Date']]
df['Days'] = df['Days'].astype('int')
df
```

```python
df1 = df.drop(df[df['Days']>48].index)
df1.info()
df1
```

```python
from collections import Counter
group_Date = df1['Line'].groupby(df1['Date'])
Line_Number = group_Date.count()
print(Line_Number)
Line_Number_count = Counter(Line_Number)
print(dict(Line_Number_count))
print(len(dict(Line_Number_count)))
```

```python
group_Line = df1['Date'].groupby(df1['Line'])
Date_Number = group_Line.count()
print(Date_Number)
Date_Number_count = Counter(Date_Number)
print(dict(Date_Number_count))
print(len(dict(Date_Number_count)))
```

```python
df_time=pd.DataFrame(Date_Number)
df_time.columns = ['Timepoint']
df_time = df_time[df_time['Timepoint']==25726]
print(df_time)
index = list(df_time.index)
print(index[::100])
```

```python
df2 = df1[df1['Line'].isin(index)]
df2.sort_values(by=['Line','Date'])
#df2.info()
```

```python
df_date = df2.groupby('Date')
total_power = df_date['Power'].sum().reset_index()
total_power['Date'] = total_power['Date'].astype('string')
total_power['Timepoint'] = [x[3:5] for x in total_power['Date']]
total_power['Timepoint'] = total_power['Timepoint'].astype('int')
print(total_power)
total_power.to_csv('CER_clean.csv')
data = pd.read_csv('CER.csv')
data = data.loc[:, ['holidays', 'Hour', 'Days', 'Month', 'Temperature', 'humiduity', 'Wind_speed']]
print(data)
```

```python
df1 = pd.read_csv('CER_clean.csv')
CER_data = pd.concat([df1, data], axis=1)
CER_data = CER_data.drop(['Date', 'Unnamed: 0'],axis=1)
CER_data.to_csv('CERdata_clean.csv')
```
