```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn import metrics
import seaborn as sns
sns.set(color_codes=True)
import matplotlib.pyplot as plt
```

```python
font1 = {
        'weight':'bold',
        'size':12,}
```

```python
import os
file_name = os.listdir(r'/home/qcraft/Documents/Fan/project/individual household')
for i in range(len(file_name)):
    file_path = r'/home/qcraft/Documents/Fan/project/individual household'+'//'+file_name[i]
    df = pd.read_csv(file_path)
    df = df.loc[:,['Power', 'Hour', 'Month', 'Days', 'Temperature', 'humiduity', 'Wind_speed']]
    corr = df.corr()
    plt.figure(figsize=(10,6))
    plt.title('correlation matrix for '+ file_name[i], font1)
    sns.heatmap(corr, cmap='Blues',annot=True)
    plt.savefig("/home/qcraft/Documents/Fan/project/individual household/correlation matrix for {}.png".format(file_name[i]))
    plt.figure(figsize=(6,4))
    plt.title('distribution of '+ file_name[i], font1)
    sns.distplot(df['Power'])
    plt.savefig("/home/qcraft/Documents/Fan/project/individual household/distribution of {}.png".format(file_name[i]))
```
