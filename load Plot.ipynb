```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

font1 = {'weight':'bold',
        'size':12,}

df = pd.read_csv('CER_data_clean.csv', index_col=0)
plt.figure(figsize=(12, 6))
plt.title('Actual Load for aggregation households', font1)
plt.xlabel('Timepoint', font1)
plt.ylabel('Magnitude', font1)
plt.plot(df['Power'])
plt.savefig('Actual Load for aggregation household')

plt.figure(figsize=(6, 4))
plt.title('Distribution of Temperature', font1)
plt.xlabel('Temperature', font1)
plt.ylabel('Density', font1)
sns.distplot(df['Temperature'])
plt.savefig('Distribution of Temperature')

plt.figure(figsize=(6, 4))
plt.title('Distribution of Humidity', font1)
plt.xlabel('Humidity', font1)
plt.ylabel('Density', font1)
sns.distplot(df['humidity'])
plt.savefig('Distribution of Humidity')

plt.figure(figsize=(6, 4))
plt.title('Distribution of Wind_speed', font1)
plt.xlabel('Wind_speed', font1) 
plt.ylabel('Density', font1)
sns.distplot(df['Wind_speed'])
plt.savefig('Distribution of Wind_speed')
```
