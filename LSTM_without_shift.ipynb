```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from keras import layers
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
import keras
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_absolute_percentage_error
```

```python
dataset = pd.read_csv('CER_data_clean.csv', header=0, index_col=0)
dataset = dataset.loc[:, ['Power', 'holidays', 'Hour', 'Days', 
                          'Month', 'Temperature', 'humiduity', 'Wind_speed']]
values = dataset.values
values = values.astype('float32')
split_point = 529 * 48
train = values[:split_point, :]
test = values[split_point:, :]
train_X, train_y = train[:, 1:], train[:, 0]
test_X, test_y = test[:, 1:], test[:, 0]

scaler = MinMaxScaler(feature_range=(0, 1))
train_X = scaler.fit_transform(train_X)
test_X = scaler.fit_transform(test_X)

train_X = train_X.reshape((train_X.shape[0], 1, train_X.shape[1]))
test_X = test_X.reshape((test_X.shape[0], 1, test_X.shape[1]))
```

```python
model = Sequential()
model.add(LSTM(100, return_sequences=True, input_shape=(train_X.shape[1], train_X.shape[2])))
model.add(LSTM(100,  return_sequences=False))
#model.add(Dropout(0.2))
model.add(Dense(1, activation="relu"))
optimizer = keras.optimizers.Adam(learning_rate=0.01)
model.compile(loss='mae', metrics=['mape'], optimizer=optimizer)
# fit network
history = model.fit(train_X, train_y, epochs=200, batch_size=32, validation_split=0.1, verbose=2, shuffle=False)
print(history.history.keys())
# plot history
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='test')
plt.title('Model loss')
plt.xlabel('Epoch')
plt.ylabel('loss')
plt.grid(True)
plt.legend()
plt.show()

y_predict = model.predict(test_X)
MAPE = mean_absolute_percentage_error(test_y,y_predict)

plt.plot(range(len(test_y)), test_y, c='r', label='actual value')
plt.plot(range(len(y_predict)), y_predict, c='b', label='prediction value')
plt.grid(True)
plt.legend()
plt.title('MAPE: %.3f' % MAPE)
plt.xlabel('Timepoint')
plt.ylabel('Power')
plt.show()
```
